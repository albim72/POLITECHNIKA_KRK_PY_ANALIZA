```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from cycler import cycler
```

```python
t=np.arange(0.0,2.0,0.01)
s = 1+np.sin(2*np.pi*t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel="czas [s]", ylabel="napięcie prądu [mV]", title="pomiar naapięcia prądu w obwodzie L45")
ax.grid()
fig.savefig('test.png')
plt.show()
```

```python
#drgania
#deklaracja danych
x1 = np.linspace(0.0,5)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)

x2 = np.linspace(0.0,2)
y2 = np.cos(2*np.pi*x2)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1)
fig.suptitle('dwa wykresy: drgania tłlumione i nietłumione')

ax1.plot(x1,y1,'o-')
ax1.set_ylabel('drgania tłumionych')

ax2.plot(x2,y2,'.-')
ax2.set_ylabel('drgania nietłumionych')
ax2.set_xlabel('czas [s]')

plt.show()
```

```python
#wykres cykli
cases = [
    None,
    8,
    (30,8),
    [16,24,30],
    [0,-1],
    slice(200,200,3),
    0.1,
    0.2,
    (0.0,0.1),
    (0.45,0.1)
]

colors = [
    '#1f77b4',
    '#ff7f0e',
    '#2ca02c',
    '#d62728',
    '#9467bd',
    '#8c564b',
    '#e377c2',
    '#7f7f7f',
    '#bcbd22',
    '#17becf',
]
```

```python
mpl.rcParams['axes.prop_cycle'] = cycler(markevery=cases,color=colors)
x = np.linspace(0,2*np.pi)
offsets = np.linspace(0,2*np.pi,11,endpoint=False) 
yy = np.transpose([np.sin(x+phi) for phi in offsets])
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.75])

for i in range(len(cases)):
    ax.plot(yy[:,i],marker='o',label=str(cases[i]))
    ax.legend(bbox_to_anchor=(1.05,1),loc='upper left',borderaxespad= 0.)

plt.title('wyświetlenie grupy wykresów cyklicznych')
plt.show()
```

```python
#hisotogram
np.random.seed(19680801)
mu = 100 #średnia rozkładu
sigma = 15 #średnie odchylenie standardowe
x = mu+sigma*np.random.randn(437)
```

```python
num_bins = 50
fig,ax = plt.subplots()
n,bins,patches = ax.hist(x,num_bins,density=True)

y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))
ax.plot(bins,y,'--')
ax.set_xlabel('wartości')
ax.set_ylabel('gęstość prawdopodobieństwa')
ax.set_title(r"Histogram wartość Q:$\mu=100$, $\sigma=15$")
fig.tight_layout()
plt.show()
```
