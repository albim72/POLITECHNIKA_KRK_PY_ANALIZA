```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.arange(0.0,2.0,0.01)
s = 1+np.sin(2*np.pi*t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel="czas [s]", ylabel="napięcie prądu [mV]", title="pomiar naapięcia prądu w obwodzie L45")
ax.grid()
fig.savefig('test.png')
plt.show()
```

```python
#wykres kołowy
labels = 'Warszawa','Kraków','Lublin','Katowice','Gdańsk'
sizes = [56,34,26,24,30]
explode = (0,0.1,0,0,0.2)
```

```python
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis('equal')
plt.show()
```

```python
#wykres słupkowy
fig,ax = plt.subplots()
marki = ['audi','toyota','mercedes','honda','kia','jeep','seat']
sprzed = [12.7,28.9,9.8,16.8,23.5,8.,28.3]
bar_labels = ['audi','toyota','mercedes','honda','kia','jeep','seat']
bar_colors = ['tab:green','tab:red','tab:green','tab:blue','tab:orange','tab:blue','tab:red']

ax.bar(marki,sprzed,label=bar_labels,color=bar_colors)
ax.set_ylabel('sprzedaż wybranych marek')
ax.set_xlabel('marki samochodów')
ax.legend(title='marki')
plt.show()
```
